### Extend
Many times people find the difference between extend and append to be unclear. So note that,\

**append: Appends object at end**\
**extend: extends list by appending elements from the iterable**

```
x = [1, 2, 3]
x.extend([4, 5])
print(x)
```

```
x = [1, 2, 3]
x.extend('ty')
print(x)
```

```
saksham=[1,2,3,4]
saksham.extend([4,5,6,'sakkiii',[10,11]])
saksham
```

# **List Comprehension**

**Q) Go through all the no.s between 1 & 100 and create a list of all the even numbers**

```
even_numbers = [x for x in range (1,101) if x%2==0]
```

**['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']**

```
mylist=['x','y','z']
required_list= [i * num for i in mylist for num in range(1,5)]
print(required_list)
```

**[[2], [3], [4], [3], [4], [5], [4], [5], [6]]**

```
required_list=[[i+n] for i in range(2,5) for n in range(3)]
print(required_list)
```
