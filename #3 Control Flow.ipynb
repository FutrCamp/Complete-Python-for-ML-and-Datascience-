### Example using if-elif-else


````
age = 1
if age < 8:
  price=0
elif age<18:
  price =5
else:
  price = 10
print("Price for Ticket is {}".format(price))
````

# **For Loops**

````
for number in range(1,11):
  print(number)
````

````
for number in [1,2,3,4,5]:
  print(number)
````

````
for letter in "abcdefghijk":
  print(letter)
````

````
vowels = 0
consonants =0

for letter in "saksham jain":
  if letter.lower() in "aeiou":
    vowels = vowels +1
  elif letter == "" :
    pass
  else : 
    consonants = consonants+1
print("There are {} vowels".format(vowels))
print("There are {} consonants".format(consonants))
````

# **While Loop**
**It repeats a block of code as long as a condition is True**

### Example: counting to 10

````
current_number = 1
````

````
while current_number<=10:
  print(current_number)
  current_number+=1
````
