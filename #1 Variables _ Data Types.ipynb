# **Variables**

### **What is a Variable?**
**Variables are used to store information to be referenced and manipulated in a computer program. They also provide a way of labeling data with a descriptive name, so our programs can be understood more clearly by the reader and ourselves.**\
*Note: Imagine them as containers that store data*

```
# Showing an output without a variable 
print("Hello World!")
```

```
# Showing the same output as above using variable
message="Hello World!"
print(message)
```

```
# It is possible to declare a variable only by assignment
x=45
x="49"
```

### **Naming Etiquettes**


1.  Use lower case variable names
2.  Use '_' for spaces
3.  letters should come before a number
         eg. variable1(Corrrect)
            1variable(incorrect)




### **For any mathematical operation.... BODMAS Rule is Followed**

# **DataTypes**
 
---


### `1) int`


*   for integers no.s  
*   immutable
*   Insertion Order Preserved : No need here
*   Indexing/Slicing : No need here
*   Duplicates : No need here
* Eg. x=7


---


### `2) float`
*   for decimals 
*   immutable
*   Insertion Order Preserved : No need here
*   Indexing/Slicing : No need here
*   Duplicates : No need here
* Eg. x=5.7
---


### `3) complex`
*   for representing complex no.s
*   immutable
*   Insertion Order Preserved : No need here
*   Indexing/Slicing : No need here
*   Duplicates :
* Eg. x=5i + 10j
---


### `4) Bool`
*   for representing Logical values  
*   immutable
*   Insertion Order Preserved : No need here
*   Indexing/Slicing : No need here
*   Duplicates : No need here
* Eg. x= True, y=False
---


### `5) str`
*   for representing sequence of characters (Multiline allowed)
*   immutable
*   Insertion Order Preserved : No need here
*   Indexing/Slicing : allowed
*   Duplicates : No need here
* Eg. x="Saksham"
---


### `6) List`
*   for representing An ordered collection of objects 
*   mutable
*   Insertion Order Preserved : Yes 
*   Indexing/Slicing : allowed
*   Duplicates : allowed
* Eg. list= [1,2,3]
---


### `7) Set`
*   an unordered collection of unique objects
*   mutable
*   Insertion Order Preserved : No 
*   Indexing/Slicing : No
*   Duplicates : No
* Eg. set={1,2,3,4}
---


### `8) frozen set`
*   Read only version of set 
*   immutable
*   Insertion Order Preserved : No need here
*   Indexing/Slicing : No
*   Duplicates : No
* Eg. froze= frozenset(set)
---


### `9) Tuple`
*   Read only version of list 
*   immutable
*   Insertion Order Preserved : yes
*   Indexing/Slicing : allowed
*   Duplicates : allowed
* Eg. tuple=(1,2,3)
---


### `10) dictionary`
*   to represent key-value pair 
*   mutable
*   Insertion Order Preserved : No
*   Indexing/Slicing : No
*   Duplicates : Key : Unique, Value: Can be duplicate
* Eg. dictionary1= {1:"Saksham",2:"Jain"}
---


### `11) bytes`
*   group of byte values  (range : (0,255)) ; for binary data representation
*   immutable
*   Insertion Order Preserved : Yes
*   Indexing/Slicing : Yes
*   Duplicates : Allowed

---


### `12) bytearray`
*   Group of Byte Values (0,255)
*   mutable
*   Insertion Order Preserved : Yes
*   Indexing/Slicing : Yes
*   Duplicates : Allowed
 
---
### `13) range`
*   representing sequence of values
*   immutable
*   Insertion Order Preserved : Yes
*   Indexing/Slicing : Yes
*   Duplicates : No

---
