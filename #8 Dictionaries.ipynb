**Dictionary has curly braces with key value pair\
Keys : Unique(Can be a string, boolian, integer, but no collection like List, tuple)\
Values: Can be anything... list,tuple,set,boolian,etc**

```
d={}
type(d)
```

```
# when a single element is present --> set
d = {1}
type(d)
```

```
d={'s':13}
print(d)
print(d['s'])
```

```
# If 2 same keys are present then the old one gets overwritten by latest value
d={'s':7,'s':108}
d['s']
print(d)
```

```
# Make a dictionary with {} and : to signify a key and a value
my_dict = {True:'value1','key2':'value2','key1':'valuedfvdfg','key1':'abc'}
my_dict
```

```
# Call values by their key
my_dict['key2']
```

```
# Dictionay inside a dictionary also possible
my_dict = {True:49,'key1':'saksham','key2':'jain','lendy':{'key1':'ss','key2':'jj'}}
print(my_dict['key2'])
print(my_dict['lendy'])
print(my_dict['lendy']['key1']) # to access list inside a list
```

```
# To one by one access the data(keys)
for i in my_dict:
    print(i)
```

```
# Dictionaries are mutable
my_dict['key2']='sakkiii'
my_dict
```

```
# To get a list of tuples of key and its value
my_dict.items()
```

```
# To get a list of all the values
my_dict.values()
```

```
# to just get the values
for i in my_dict.values():
    print(i)
```

## A few Dictionary Methods

There are a few methods we can call on a dictionary. Let's get a quick introduction to a few methods:

```
# Create a typical dictionary
d = {'key1':1,'key2':2,'key3':3}
```

```
# Method to return a list of all keys 
f=d.keys()
print(f)
list(f)[1]
```

```
# Method to grab all valu
d.values()
```

```
# Method to return tuples of all items  (we'll learn about tuples soon)
d.items()
```

## Dictionary Comprehensions

```
# to print a key value pair for a nuner and its square
{i: i**2 for i in range(10)}
```

```
{i:i+1 for i in range(13)}
```
