*args

```
def f1(*n):
  print(n)
```

```
f1()
f1(1)
f1(10)
f1(1,10)
f1(1,2,3,4,5)
```

**kwargs

```
def f1(**kwargs):
  print(kwargs)
```

```
f1()
f1(name='python',roll_no=1)
```

```
def f1(*args,**kwargs,name):
  print(args)
  print(kwargs)
```

```
f1(10,2,4,6,8,A=10,B=30)
```

```
def f1(**kwargs,*args):
  print(kwargs)
  print(args)
```
