# **Tuples**
It is a Datatype similar to lists but they are immutable.

Declaring Tuples

```
our_tuple1 = 1,2,3,"A","B","C"
our_tuple2 = (1,2,3,"A","B","C")
```

```
print(type(our_tuple1))
print(type(our_tuple2))
```

```
our_tuple1[0:3]
```

```
our_tuple2[0:4]
```

Tuple are immutable

```
our_list=[1,2,3,45,5,64,5]
print(our_list)
our_list[1] = 100000
print(our_list)
```

```
our_tuple=  (1,2,3,45,5,64,5)
print(our_tuple)
our_tuple[1] = 1000000
print(our_tuple)
```

List into tuple

```
my_list= [12,34,23,545]
type(my_list)
```

```
print(type(tuple(my_list)))
```

```
my_tuple= tuple(my_list)
print(my_tuple)
```

**Random Stuff**
`V. imp`

```
D,E,F = [1,2,3]
print(D)
print(E)
print(F)
```

# **Sets**

Sets are an unordered collection of `*unique*` elements which can be constructed using the set() function. 

Let's go ahead and create a set to see how it works.\
**To remove duplicate data, use set**

```
mylist = list("Saksham Jain")
```

```
print(mylist)
```

```
myset= set(mylist)
```

```
print(myset)
```

 `Note` If we pass an integer in a list() or set() --> `error`\
because they are expecting an iterable element


```
set(1)
```

```
list(1)
```

### `NOTE` Iterable data means u can extract information out of that data
eg. string\
eg of non iterable : integer

```
set([1,2,3,4,5,6,7,8,8,888,99999])
```

### tuples inside a set


```
set((2,3,4,4,4,))
```

### Adding

```
# We add to sets with the add() method
x=set([49])
x.add(3)
x
```
